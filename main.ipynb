```python
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup

def scrape_page(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    rows = soup.select("tbody.Crom_body__UYOcU tr")
    data = []
    for row in rows:
        row_data = [td.get_text() for td in row.find_all("td")]
        data.append(row_data)
    return data

def main():
    # Set up Selenium webdriver
    driver = webdriver.Chrome()

    # Load the NBA stats page
    driver.get("https://www.nba.com/stats/players/boxscores?dir=D&sort=GDATE")

    data = []

    while True:
        data += scrape_page(driver)
        print("Scraped data from current page.")
        
        # Check if the next button is present
        next_buttons = driver.find_elements_by_css_selector("button[data-track='click'][data-type='controls'][data-pos='next']")
        if not next_buttons:
            print("No next button found. Exiting loop.")
            break
        
        # Click the first next button found
        next_buttons[0].click()
        print("Clicked next page button.")
        
        # Wait for the page to load after clicking
        time.sleep(5)  # Adjust this timeout as needed

    # Create a DataFrame from the list of lists
    columns = [
        "Player", "Team", "Opponent", "Date", "Result", "Minutes", "Points", 
        "FGM", "FGA", "FG%", "3PM", "3PA", "3P%", "FTM", "FTA", "FT%", 
        "OREB", "DREB", "REB", "AST", "TO", "STL", "BLK", "PF", "+/-", "SPI"
    ]
    df = pd.DataFrame(data, columns=columns) 
    df.to_csv("nba_player_boxscores_2023-2024.csv", index=False)

    # Close the Selenium webdriver
    driver.quit()

# Call the main function
main()
```
